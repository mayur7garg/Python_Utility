# Discrete Fourier Transform

Reference: [YouTube: Coding Adventure: Sound (and the Fourier Transform)](https://youtu.be/iA6wRgwl7k0)

## Imports

```python
import dft
```

## Original signal

```python
DURATION = 4
SAMPLE_RATE = 32
FREQ_COUNT = 3
MAX_FREQ = 10
MAX_AMP = 1.2

sig, freqs, amps, phases = dft.generate_random_signal(DURATION, SAMPLE_RATE, FREQ_COUNT, MAX_FREQ, MAX_AMP, print_summary = True)
dft.plot_signal(sig)
```

## Fourier transform of a single frequency

```python
freq_to_check = 1

freq_result = dft.check_freq(freq_to_check * DURATION, sig, dur = DURATION, plot = True)
print(f"Actual freq: {freq_result.freq:.3f} Hz")
print(f"Amplitude: {freq_result.amp:.3f}")
print(f"Phase: {freq_result.phase:.3f} radians")
```

## Comparison plot in frequency domain

```python
dft.plot_amp_comparison(
    freqs, 
    amps, 
    dft.check_freq_range(range(((DURATION * SAMPLE_RATE) // 2) + 1), sig, DURATION)
)
```
